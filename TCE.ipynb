```python
from unicodedata import normalize
import pandas as pd
```

```python
options = input('Escolha entre DESPESAS ou RECEITAS: ')
city = input('Digite o nome da cidade: ')
year = input('Digite o Ano: ')
month = input('Digite o mês: ')
```

```python
url = f'https://transparencia.tce.sp.gov.br/api/xml/{options}/{city}/{year}/{month}'
df = pd.read_xml(url)

# Definindo estrutura de condições 

if options == "despesas":
  colunas =  df[['mes', 'nm_fornecedor', 'dt_emissao_despesa', 'vl_despesa']]
  display(colunas)
else: 
  colunas = [['mes', 'ds_fonte_recurso', 'ds_alinea', 'vl_arrecadacao']]
  display(colunas)
```
